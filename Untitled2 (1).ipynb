2. write a program that creates a databases named phonebook.db. the database should have a table named Entries, with columns for a persons name and phone number. next, write a CRUD application that lets the user add rows to the Entries table, look up a persons phone number, change a persons phone number, and delete specified rows.

```python
import sqlite3
import tkinter as tk
from tkinter import Label, Entry, Button, messagebox

# function to create the database and table
def create_database():
    connection = sqlite3.connect("phonebook.db")
    cursor = connection.cursor()

    # create the Entries table
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS Entries (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT NOT NULL,
            phone_number TEXT NOT NULL
        )
    ''')

    connection.commit()
    connection.close()

# function to add an entry to the database
def add_entry():
    name = name_entry.get()
    phone_number = phone_entry.get()

    if name and phone_number:
        connection = sqlite3.connect("phonebook.db")
        cursor = connection.cursor()

        cursor.execute('''
            INSERT INTO Entries (name, phone_number) VALUES (?, ?)
        ''', (name, phone_number))

        connection.commit()
        connection.close()

        messagebox.showinfo("Success", "Entry added successfully.")
        clear_entries()
    else:
        messagebox.showerror("Error", "Name and phone number are required.")

# function to look up a person's phone number
def lookup_entry():
    name = name_entry.get()

    if name:
        connection = sqlite3.connect("phonebook.db")
        cursor = connection.cursor()

        cursor.execute('''
            SELECT phone_number FROM Entries WHERE name = ?
        ''', (name,))

        result = cursor.fetchone()

        connection.close()

        if result:
            messagebox.showinfo("Phone Number", f"The phone number for {name} is {result[0]}.")
        else:
            messagebox.showwarning("Not Found", f"No entry found for {name}.")
    else:
        messagebox.showerror("Error", "Name is required.")

# function to update a person's phone number
def update_entry():
    name = name_entry.get()
    new_phone_number = phone_entry.get()

    if name and new_phone_number:
        connection = sqlite3.connect("phonebook.db")
        cursor = connection.cursor()

        cursor.execute('''
            UPDATE Entries SET phone_number = ? WHERE name = ?
        ''', (new_phone_number, name))

        connection.commit()
        connection.close()

        messagebox.showinfo("Success", "Phone number updated successfully.")
        clear_entries()
    else:
        messagebox.showerror("Error", "Name and new phone number are required.")

# function to delete a specified row
def delete_entry():
    name = name_entry.get()

    if name:
        confirmation = messagebox.askyesno("Confirm Deletion", f"Do you want to delete the entry for {name}?")

        if confirmation:
            connection = sqlite3.connect("phonebook.db")
            cursor = connection.cursor()

            cursor.execute('''
                DELETE FROM Entries WHERE name = ?
            ''', (name,))

            connection.commit()
            connection.close()

            messagebox.showinfo("Success", "Entry deleted successfully.")
            clear_entries()
    else:
        messagebox.showerror("Error", "Name is required.")

# function to clear entry fields
def clear_entries():
    name_entry.delete(0, tk.END)
    phone_entry.delete(0, tk.END)

# create the database and table
create_database()

# create the main window
root = tk.Tk()
root.title("Phonebook Application")

# create and place widgets in the window
name_label = Label(root, text="Name:")
name_label.grid(row=0, column=0, padx=10, pady=10)

name_entry = Entry(root)
name_entry.grid(row=0, column=1, padx=10, pady=10)

phone_label = Label(root, text="Phone Number:")
phone_label.grid(row=1, column=0, padx=10, pady=10)

phone_entry = Entry(root)
phone_entry.grid(row=1, column=1, padx=10, pady=10)

add_button = Button(root, text="Add Entry", command=add_entry)
add_button.grid(row=2, column=0, columnspan=2, pady=10)

lookup_button = Button(root, text="Lookup Entry", command=lookup_entry)
lookup_button.grid(row=3, column=0, columnspan=2, pady=10)

update_button = Button(root, text="Update Entry", command=update_entry)
update_button.grid(row=4, column=0, columnspan=2, pady=10)

delete_button = Button(root, text="Delete Entry", command=delete_entry)
delete_button.grid(row=5, column=0, columnspan=2, pady=10)

clear_button = Button(root, text="Clear Entries", command=clear_entries)
clear_button.grid(row=6, column=0, columnspan=2, pady=10)

# start the GUI event loop
root.mainloop()
```
